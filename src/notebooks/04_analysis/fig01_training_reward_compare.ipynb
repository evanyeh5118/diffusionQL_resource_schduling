```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

import sys
sys.path.insert(0, '../../../../')

import numpy as np
import matplotlib.pyplot as plt
import pickle

from fig01_helper import plot_two_groups, load_reward_curves
```

```python


# NOTE: Only config_0 is available in the data directory
# If you need to compare with config_1, please ensure those model files are generated first
# by running the training notebooks for config_1
paths_forward = [f'config_0', f'config_0/without_bc']
paths_backward = [f'config_0', f'config_0/without_bc']  # Updated: config_1 doesn't exist yet
idxs_forward = range(1)
idxs_backward = range(1)

models_dir = "../../../data/results/dql"

# Load data for both sets
print("Loading forward data...")
reward_curves_forward = load_reward_curves(paths_forward, idxs_forward, models_dir=models_dir)

print("Loading backward data...")
reward_curves_backward = load_reward_curves(paths_backward, idxs_backward, models_dir=models_dir)

# Prepare data for plotting
reward_curves_forward_list = [reward_curves_forward[lb] for lb in paths_forward]
reward_curves_backward_list = [reward_curves_backward[lb] for lb in paths_backward]

# Labels for forward curves
labels_forward = ['w/ Expert', 'w/o Expert']
labels_final_forward = ['w/ Expert', 'w/o Expert']

# Labels for backward curves
labels_backward = ['Config 10 (Avg.)', 'Without BC (Avg.)']
labels_final_backward = ['Expert (Final)', 'Without BC (Final)']


# Plot both sets on the same figure
print("Creating combined plot...")
fig = plot_two_groups(
    curves_list1=reward_curves_forward_list,
    labels1=labels_forward,
    curves_list2=reward_curves_backward_list,
    labels2=labels_backward,
    figsize=(11, 3),
    title1="8 Users",
    title2="20 Users",
    title_fontsize=25,
    label_fontsize=22,
    tick_fontsize=20,
    legend_fontsize=20
)

# Save the figure
output_path = 'combined_training_curves.png'
plt.savefig(output_path, dpi=300, bbox_inches='tight')
print(f"Figure saved as: {output_path}")


#targetFolder = "C:/Users/Ye/Documents/YuYeh_Documents/L2S/Conference_Writing/ICC_diffusion_policy/diffusion_model/Figures"
#fig.savefig(os.path.join(targetFolder, "train_curve_haptic.png"), dpi=300, bbox_inches='tight')
```
